# Data Cleaning
---

```python
df.fillna(np.nan, axis=0, inplace=True)
```

```python
#print columns missing values count and which columns have missed values
missing_data = df.isnull()
cols_na = []
for col in missing_data.columns.values.tolist():
    true_data = missing_data[col].value_counts().to_frame().reset_index()
    if len(true_data)>1:
        print(true_data[true_data['index']==True])
        cols_na.append(true_data.columns[1])
        print("")
        print("--------------------------------")
        
cols_na
```

```python
df[cols_na].dtypes
```

```python
df['Preferred Foot'].replace(np.nan,'Right', inplace=True)
```

```python
df['Work Rate'].replace(np.nan,'Medium/ Medium', inplace=True)
df['Body Type'].replace(np.nan,'Normal', inplace=True)
df['Real Face'].replace(np.nan,'No', inplace=True)
df['Club'].dropna(axis=0, inplace=True)
df['Position'].dropna(axis=0, inplace=True)
df.drop('Loaned From', axis=1, inplace=True)
```

```python
df['Joined'] = pd.to_datetime(df['Joined'])
df['Joined'] = pd.to_numeric(df['Joined'])
mean_joined_data = pd.pivot_table(df, values='Joined', index=['Club'], aggfunc=np.mean)
mean_joined_data = mean_joined_data.reset_index()

df['Joined'] = df.Joined.fillna(mean_joined_data.Joined, axis=0)
df['Joined'] = pd.to_datetime(df['Joined'])
df['Joined'].isnull().sum()
```

```python
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'])
df['Contract Valid Until'] = pd.to_numeric(df['Contract Valid Until'])
mean_Contract_Valid_Until_data = pd.pivot_table(df, values='Contract Valid Until', index=['Club'], aggfunc=np.mean)
mean_Contract_Valid_Until_data['Joined'] = pd.to_datetime(mean_Contract_Valid_Until_data['Contract Valid Until'])
mean_Contract_Valid_Until_data = mean_Contract_Valid_Until_data.reset_index()

df['Contract Valid Until'].replace(np.nan, mean_Contract_Valid_Until_data['Contract Valid Until'], inplace=True)
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'])
df['Contract Valid Until'].isnull().sum()
```

```python
df['Height'] = pd.to_numeric(df.Height.str.replace("'", "."))
df['Weight'] = pd.to_numeric(df.Weight.str.replace("lbs", ""))
```

```python
df['Weight'].replace(np.nan, np.mean(df['Weight']), inplace=True)
df['Height'].replace(np.nan, np.mean(df['Height']), inplace=True)
print(df['Weight'].head())
print(df['Height'].head())
```

```python
df['Height'].isnull().sum()
```

```python
# change the column Value to numbers from strings
for i in range(len(df.Value.values[::-1])):
    if "M" in df.Value[i]:
        df.Value.values[i] = df.Value.values[i].replace("M","")
        df.Value.values[i] = df.Value.values[i].replace("€","")
        df.Value.values[i] = pd.to_numeric(df.Value.values[i])*1000000
    elif "K" in df.Value[i]:
        df.Value.values[i] = df.Value.values[i].replace("K","")
        df.Value.values[i] = df.Value.values[i].replace("€","")
        df.Value.values[i] = pd.to_numeric(df.Value.values[i])*1000
        
df.Value
```

```python
#change wage columns to numeric
df.Wage = pd.to_numeric(df.Wage.str.replace("[€,K]", ""))*1000
```

```python
#Change the body type of the players to the proper type
df['Body Type'].value_counts()
```

```python
#change the body type to proper ones
df.loc[df['Body Type'] == 'Shaqiri','Body Type'] = 'Stocky'
df.loc[df['Body Type'] == 'PLAYER_BODY_TYPE_25','Body Type'] = 'Normal'
df.loc[df['Body Type'] == 'Neymar','Body Type'] = 'Normal'
df.loc[df['Body Type'] == 'Messi','Body Type'] = 'Normal'
df.loc[df['Body Type'] == 'C. Ronaldo','Body Type'] = 'Normal'
df.loc[df['Body Type'] == 'Courtois','Body Type'] = 'Normal'
df.loc[df['Body Type'] == 'Akinfenwa','Body Type'] = 'Stocky'
df['Body Type'].value_counts()
```

```python
df['Release Clause'].replace(np.nan, 'M', inplace=True)
df['Release Clause'].isna().sum()
```

```python
# change the data type of the release clause
for i in range(len(df['Release Clause'].values)):
    if "M" in df['Release Clause'].values[i]:
        df['Release Clause'].values[i] = df['Release Clause'].values[i].replace("M","")
        df['Release Clause'].values[i] = df['Release Clause'].values[i].replace("€","")
        df['Release Clause'].values[i] = pd.to_numeric(df['Release Clause'].values[i])*1000000
    elif "K" in df['Release Clause'].values[i]:
        df['Release Clause'].values[i] = df['Release Clause'].values[i].replace("K","")
        df['Release Clause'].values[i] = df['Release Clause'].values[i].replace("€","")
        df['Release Clause'].values[i] = pd.to_numeric(df['Release Clause'].values[i])*1000
        
df['Release Clause'].replace(np.nan, np.mean(df['Release Clause']), inplace=True)
```
